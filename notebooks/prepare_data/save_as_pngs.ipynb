```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os

output_dir = "/store1/candy/overlaid_frames_preCropNet_cropped" # save directory for pngs
path_h5 = "/store1/candy/prj_unet2d/data/preCropNet_cropped.h5" # hdf5 with image-label pairs

# output_dir = "/store1/candy/overlaid_frames_immob_preCropNet_cropped"
# path_h5 = "/store1/candy/prj_unet2d/data/immob_preCropNet_cropped.h5"

# output_dir = "/store1/candy/overlaid_frames_postCropNet"
# path_h5 = "/store1/candy/prj_unet2d/data/postCropNet.h5"

# output_dir = "/store1/candy/overlaid_frames_immob_postCropNet"
# path_h5 = "/store1/candy/prj_unet2d/data/immob_postCropNet.h5"

os.makedirs(output_dir, exist_ok=True)

# Open the h5 file
with h5py.File(path_h5, 'r') as file:
    # Loop through all keys (frames) in the file
    for i in file.keys():
        # Load image and label
        img_raw = file[i]['img'][()]
        img_label = file[i]['label'][()]
        
        # Create a figure
        plt.figure(figsize=(10, 8))
        
        # Display the raw image
        plt.imshow(img_raw, cmap='gray')
        
        # Overlay the label with transparency
        # Using a colormap that contrasts well with grayscale (like 'jet' or 'viridis')
        plt.imshow(img_label, alpha=0.4, cmap='viridis')
        
        plt.title(f"Frame {i}")
        plt.axis('off')  # Hide axes for cleaner image
        
        # Save as PNG
        output_path = os.path.join(output_dir, f"frame_{i}.png")
        plt.savefig(output_path, bbox_inches='tight', dpi=150)
        plt.close()  # Close the figure to free memory
        
        print(f"Saved frame {i} to {output_path}")

print("All frames saved successfully!")
```
